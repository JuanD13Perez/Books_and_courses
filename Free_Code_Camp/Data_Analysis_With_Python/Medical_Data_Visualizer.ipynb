```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Import data
df = pd.read_csv('medical_examination.csv')

# Add 'overweight' column
df['overweight'] = np.where(np.array((df['weight']/((df['height']/100)**2))) > 25 , 1,0)

# Normalize data by making 0 always good and 1 always bad. If the value of 'cholesterol' or 'gluc' is 1, make the value 0.
# If the value is more than 1, make the value 1.
dict_normalize = {1:0 , 2:1, 3:1}
df['gluc'] = df['gluc'].map(dict_normalize)
df['cholesterol'] =df['cholesterol'].map(dict_normalize)

# Draw Categorical Plot
def draw_cat_plot():
    # Create DataFrame for cat plot using `pd.melt` using just the values from 'cholesterol', 'gluc', 'smoke', 'alco', 'active', and 'overweight'.
    df_cat = pd.melt(df[['cardio','cholesterol', 'gluc', 'smoke', 'alco', 'active', 'overweight']], id_vars='cardio')
    df_cat['total'] = 1

    # Group and reformat the data to split it by 'cardio'. Show the counts of each feature. You will have to rename one of the columns
    # for the catplot to work correctly.

    df_gr= df_cat.groupby(['cardio','variable','value']).count()
    df_gr = df_gr.reset_index()
    

    # Draw the catplot with 'sns.catplot()'
    # sns.catplot(data = df_gr, x = 'variable', y='total',col ='cardio',hue='value', kind= 'bar')
    


    # Get the figure for the output
    fig = sns.catplot(data = df_gr, x = 'variable', y='total',col ='cardio',hue='value', kind= 'bar')
    plt.show()


    # Do not modify the next two lines
    fig.savefig('catplot.png')
    return fig


# Draw Heat Map
def draw_heat_map():
    h025 = df['height'].quantile(0.025)
    h975 = df['height'].quantile(0.975)

    w025 = df['weight'].quantile(0.025)
    w975 = df['weight'].quantile(0.975)


    df_heat = df[((df['ap_lo'] <= df['ap_hi'])) & ((df['height'] >= h025) & (df['height'] <= h975)) & ((df['weight'] >= w025) & (df['weight'] <= w975))]

    # Calculate the correlation matrix
    corr = df_heat.corr()

    # Generate a mask for the upper triangle
    mask = np.zeros_like(corr, dtype=np.bool)
    mask[np.triu_indices_from(mask)] = True



    # Set up the matplotlib figure
    fig, ax = plt.subplots(figsize=(15,7))


    # Draw the heatmap with 'sns.heatmap()'
    sns.heatmap(corr,mask = mask, annot= True, fmt='0.1f')

    # Do not modify the next two lines
    fig.savefig('heatmap.png')
    return fig
```

```python
draw_cat_plot()
```

```python
draw_heat_map()
```
